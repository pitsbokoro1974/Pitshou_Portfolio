```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Unit06 wine.csv')

# Select only feature columns for clustering

df.head()
```

```python
X = df.drop(labels=['Wine'], axis=1)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python

clusters = kmeans.labels_  # array of 0,1,2 assigning each point to a cluster
df['cluster'] = clusters
print(df.head())
```

```python
print(pd.crosstab(df['Wine'], df['cluster']))
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

plt.scatter(X['Alcohol'], X['Malic.acid'], c=clusters, cmap='viridis')
plt.xlabel('Alcohol')
plt.ylabel('Malic.acid')
plt.show()
```
