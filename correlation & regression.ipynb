```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np

x = pd.read_csv('Unit04 Global_GDP.csv')
y = pd.read_csv('Unit04 Global_Population.csv')


y.head()
```

```python
#drop the missing values 
x.dropna(axis=0, inplace=True)
y.dropna(axis=0, inplace=True)
```

```python
#drop uncommon identifiers
x = x.drop(labels=['Indicator Name', 'Indicator Code'], axis=1)
y = y.drop(labels=['Series Name', 'Series Code'], axis=1)
```

```python
x = x.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='GDP' )
y = y.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Population')
```

```python
x.head()
```

```python
y.head()
```

```python
corr_data = pd.merge(x, y, how="inner")
corr_data = corr_data[corr_data['Year'].astype(int) >= 2001]

mask = ~corr_data['Population'].str.replace(',', '', regex=False).str.strip().str.fullmatch(r'\d+(\.\d+)?')

corr_data['Population'] = pd.to_numeric(corr_data['Population'].str.replace(',', '', regex=False), errors='coerce')
```

```python
corr_data.head()


corr_data['Population'] = pd.to_numeric(corr_data['Population'])
corr_data.drop(labels=['Country Name', 'Country Code', 'Year'], axis=1, inplace=True)
corr_data.head()
```
