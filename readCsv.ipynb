```python
import pandas as pd
x = pd.read_csv("Unit02 train (1).csv")
x
```

```python
x.isnull().sum()
```

```python
(x.isnull().sum() / len(x)) * 100
```

```python
x[x.isnull().any(axis=1)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(x.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```
